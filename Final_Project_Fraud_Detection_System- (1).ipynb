#### since our dataset is imbalanced so our model is overfitted giving the bad Accuracy -
- Lets handle this imbalanced Dataset  and trying to make it balanced-

### We have two techniques to Handle imbalance Dataset-
- **Under Sampling** - we are randomly removing rows from majority class to match them with minority class-
- **Oversamppling** - balance class distribution by randomly increasing minority class rows with synthesizing new minority data points from existing instances (generates virtual record by linear interpolation without generating duplicates)

### Let us implement UnderSampling-

```python
fair = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
fair.shape
```

```python
fraud.shape
```

```python
fair_sample = fair.sample(n=473)
```

```python
fair_sample.shape
```
